# Plotting Loss

**Compute loss for a given Dataset:** (Grid Search approach):   (brute force approach is inefficient, we will use gradient decent later.)

```
def sigmoid_function(w,x,b):
  return 1/(1+np.exp(-(w*x+b)))
```

```
# We are given with the input and the output, we need to find the corresponding weights and b. This is a typical ML task.
w_unknown= 0.5    #initialized to a value
b_unknown= 0.25   #initialized to a value
x= np.random.random(25)*20-10
y= sigmoid_function(w_unknown,x,b_unknown) #w_unknown and b_unknown are values not known to us, we want to find them;
```

```
plt.plot(x,y,'*')
plt.xlabel('x')
plt.ylabel('y');     
# for these x and y values, can we find the corresponding w_unknown and b_unknown?
```

```
def calculate_loss(x,y,w_est,b_est):
  loss=0
  for i,j in zip(x,y):
    loss+= np.square((j- sigmoid_function(w_est,i,b_est)))
  return loss
```

```
w= np.linspace(-1,1,100)    #someone will have to tell us, that the values of w will lie between -1 to 1 (later we would do using gradient)
b= np.linspace(-1,1,100)    # these are hypothesis

W,B= np.meshgrid(w,b)       # for each such hypothesis, we would compute the loss funciton
Loss= np.zeros(W.shape)
```

```
for i in range(W.shape[0]):     #we could also have chosen B.shape, since they have same shape.
  for j in range(W.shape[1]):
    Loss[i,j]= calculate_loss(x,y,W[i,j],B[i,j])
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(W, B, Loss, cmap='inferno')
ax.set_xlabel('Weight')
ax.set_ylabel('Bias')
ax.set_zlabel('Loss')
plt.tight_layout();
```

Note: The two plots are different; One of them gives the loss function (for all w and b) whereas the other one gives the output given 2 inputs (for some given w and b).

**From the above curve, we can see that the negative value of "w" is not good, because in the plot between the x and y, it will give a completely reverse plot. We want the values of **  w and b **to be such that loss is minimum.** So, we would want to refine the search space:

```
w= np.linspace(0,2,101)    #We have changed the value of w because we don't want the negative part
b= np.linspace(-1,1,101)    

W,B= np.meshgrid(w,b)       
Loss= np.zeros(W.shape)

for i in range(W.shape[0]):
  for j in range(W.shape[1]):
    Loss[i,j]= calculate_loss(x,y,W[i,j],B[i,j])
    
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(W, B, Loss, cmap='inferno')
ax.set_xlabel('Weight')
ax.set_ylabel('Bias')
ax.set_zlabel('Loss')
plt.tight_layout();
```

But we don't even need these evaluations, since we want the values of w and b corresponding to minimum loss.

```
i,j=np.unravel_index(Loss.argmin(), Loss.shape)   #argmin will give the index after
                                                  #flattening out in an array, then unravel_index will 
                                                  #give the index of loss corresponding to min.
print(i,j)
```

```
print(W[i,j],B[i,j])   #these are the weight and bias that gives minimum Loss/error.
```

# Class for Sigmoid Neuron

```
class SigmoidNeuron:
  def __init__(self):
    self.w= None
    self.b= None
  
  def perceptron(self,x):
    return np.dot(self.w,x)+self.b
  
  def sigmoid(self,x):
    return 1.0/(1+np.exp(-x))
  
  def grad_w(self,x,y):
    y_pred= self.sigmoid(self.perceptron(x))
    return (y_pred-y)*(1-y_pred)*y_pred*x
  
  def grad_b(self,x,y):
    y_pred= self.sigmoid(self.perceptron(x))
    return (y_pred-y)*(1-y_pred)*y_pred
  
  def fit(self,X,Y,epochs=1,learning_rate=1,initialize= True):
    if initialize:
      self.w= np.random.randn(1,X.shape[1])   #w will have the same number of cols as X has.
      self.b= 0
    
    for i in range(epochs):
      dw= 0
      db= 0
      
      for x,y in zip(X,Y):
        dw+= self.grad_w(x,y)
        db+= self.grad_b(x,y)
      self.w -= learning_rate*dw
      self.b -= learning_rate*db
      
        
  
```

**Visualizing the decision boundary:**

```
sn = SigmoidNeuron()
```

```
def plot_sn(X, Y, sn, ax):
  X1 = np.linspace(-10, 10, 100)   # axis 1 chosen between -10,10
  X2 = np.linspace(-10, 10, 100)   # axis 2 chosen between -10,10
  XX1, XX2 = np.meshgrid(X1, X2)   # rectangular grid is made from the two
  YY = np.zeros(XX1.shape)         
  for i in range(X2.size):         # Notice the order
    for j in range(X1.size):         
      val = np.asarray([X1[j], X2[i]])     #val contains the coordinate of the first point
      YY[i, j] = sn.sigmoid(sn.perceptron(val))   # YY basically contains the y_hat value.
  ax.contourf(XX1, XX2, YY, cmap=my_cmap, alpha=0.6)   # Contour map for Sigmoid function is plotted
  
  # Now plotting the X and Y on the same contour plot:
  ax.scatter(X[:,0], X[:,1],c=Y, cmap=my_cmap)         # Scatter plot for X and Y.
  ax.plot()
```

```
sn.fit(X, Y, 1, 0.05, True)
N = 30
plt.figure(figsize=(10, N*5))
for i in range(N):
  print(sn.w, sn.b)
  ax = plt.subplot(N, 1, i + 1)
  plot_sn(X, Y, sn, ax)
  sn.fit(X, Y, 1, 0.5, False)
```

## Test-Train split

**Should we split first and then do the standardisation or the opposite?**

**We should split first and then standardize/normalize.** 
 If we take the mean and variance of the whole dataset we will be introducing future information into the training explanatory variables (i.e. the mean and variance). 
 
 Therefore, we should perform feature normalisation over the training data. Then perform nomralisation on testing instances as well, but this time using the mean and variance of training explanatory variables. _In this way, we can test and evaluate whether our model can generalize well to new, unseen data points._
 
 That's why sklearn privide fit and then transform, or fit_transform if we are doing the fit and transform on the same data.

```
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test= train_test_split(X,y,random_state=0,stratify=y_binarised)

#stratify is done on the binarised values
```

```
print(X_train.shape,X_test.shape)
```

```
scaler= StandardScaler()
X_scaled_train= scaler.fit_transform(X_train)  #fit and transform X_train
X_scaled_test= scaler.transform(X_test)  # only transform from the fit obtained from X_train
```

Warnings are because we are having some float values and some ints; so all ints are getting converted into floats.

**Now, with standardscaler, we are converting the training data into 0 mean and 1 std. But what about y_train?**

Will it work, if we standardize these values? We know that our sigmoid Neuron takes computes values between 0 and 1. So stanadrdizing 'y' would not help. We need to transform it such that our values lie between 0 and 1.

## Fitting Data:

```
class SigmoidNeuron:
  def __init__(self):
    self.w= None
    self.b= None
  
  def perceptron(self,x):
    return np.dot(self.w,x)+self.b
  
  def sigmoid(self,x):
    return 1.0/(1+np.exp(-x))
  
  def grad_w(self,x,y):
    y_pred= self.sigmoid(self.perceptron(x))
    return (y_pred-y)*(1-y_pred)*y_pred*x
  
  def grad_b(self,x,y):
    y_pred= self.sigmoid(self.perceptron(x))
    return (y_pred-y)*(1-y_pred)*y_pred
  
  def fit(self,X,Y,epochs=1,learning_rate=1,initialize= True):
    if initialize:
      self.w= np.random.randn(1,X.shape[1])   #w will have the same number of cols as X has.
      self.b= 0
    
    for i in range(epochs):
      dw= 0
      db= 0
      
      for x,y in zip(X,Y):
        dw+= self.grad_w(x,y)
        db+= self.grad_b(x,y)
      self.w -= learning_rate*dw
      self.b -= learning_rate*db
      
        
  def predict(self,X):
    Y_pred= []
    for x in X:
      y_pred= self.sigmoid(self.perceptron(x))
      Y_pred.append(y_pred)
    return np.array(Y_pred)
  
```

```
from sklearn.metrics import accuracy_score

sig= SigmoidNeuron()

sig.fit(X_scaled_train,y_scaled_train,epochs=5000,learning_rate=0.01) #later we will come to how to choose epoch and learning rate
```

```
y_pred_train= sig.predict(X_scaled_train)  #these give ground truth
y_pred_test= sig.predict(X_scaled_test)
```

```
y_pred_binarised_train= (y_pred_train> scaled_threshold).astype('int').ravel()  #binarising
y_pred_binarised_test = (y_pred_test > scaled_threshold).astype('int').ravel() 
```

```
accuracy_train= accuracy_score(y_pred_binarised_train,y_binarised_train)
accuracy_test= accuracy_score(y_pred_binarised_test,y_binarised_test)
```

```
print(accuracy_train,accuracy_test)
```

## Plotting Loss

```
from sklearn.metrics import mean_squared_error

class SigmoidNeuron:
  def __init__(self):
    self.w= None
    self.b= None
  
  def perceptron(self,x):
    return np.dot(x,self.w.T)+self.b
  
  def sigmoid(self,x):
    return 1.0/(1+np.exp(-x))
  
  def grad_w(self,x,y):
    y_pred= self.sigmoid(self.perceptron(x))
    return (y_pred-y)*(1-y_pred)*y_pred*x
  
  def grad_b(self,x,y):
    y_pred= self.sigmoid(self.perceptron(x))
    return (y_pred-y)*(1-y_pred)*y_pred
  
  def fit(self,X,Y,epochs=1,learning_rate=1,initialize= True,display_loss=False):
    if initialize:
      self.w= np.random.randn(1,X.shape[1])   #w will have the same number of cols as X has.
      self.b= 0
    
    if display_loss:
      loss= {}
    
    for i in range(epochs):
      dw= 0
      db= 0
      
      for x,y in zip(X,Y):
        dw+= self.grad_w(x,y)
        db+= self.grad_b(x,y)
      self.w -= learning_rate*dw
      self.b -= learning_rate*db
      
     
      if display_loss:
        Y_pred= self.sigmoid(self.perceptron(X))
        loss[i]= mean_squared_error(Y_pred,Y)
    
    if display_loss:
      plt.plot(loss.values())
      plt.xlabel('Epochs')
      plt.ylabel('Mean Squared Error')
    
  def predict(self,X):
    Y_pred= []
    for x in X:
      y_pred= self.sigmoid(self.perceptron(x))
      Y_pred.append(y_pred)
    return np.array(Y_pred)
  
```

```
from sklearn.metrics import accuracy_score

sig= SigmoidNeuron()

sig.fit(X_scaled_train,y_scaled_train,epochs=5000,learning_rate=0.01,display_loss=True) #later we will come to how to choose epoch and learning rate
```

**Fine tuning the hyperparameters, by looking at the loss plots is very important task in Deep learning.** As we can see here, that after around 1200, the errors is not varying much; so to improve accuracy, we may want to decrese the epochs and increase the learning rate. **Increasing learning rate assures that in the region where the loss plot is flattening out, we are giving the algorithm some more chance in decreasing the loss.**

```
sig.fit(X_scaled_train,y_scaled_train,epochs=2000,learning_rate=0.015,display_loss=True)
```

## Adding a progress bar:

A minor change made in the epochs:

```
from tqdm import tqdm_notebook

from sklearn.metrics import mean_squared_error

class SigmoidNeuron:
  def __init__(self):
    self.w= None
    self.b= None
  
  def perceptron(self,x):
    return np.dot(x,self.w.T)+self.b
  
  def sigmoid(self,x):
    return 1.0/(1+np.exp(-x))
  
  def grad_w(self,x,y):
    y_pred= self.sigmoid(self.perceptron(x))
    return (y_pred-y)*(1-y_pred)*y_pred*x
  
  def grad_b(self,x,y):
    y_pred= self.sigmoid(self.perceptron(x))
    return (y_pred-y)*(1-y_pred)*y_pred
  
  def fit(self,X,Y,epochs=1,learning_rate=1,initialize= True,display_loss=False):
    if initialize:
      self.w= np.random.randn(1,X.shape[1])   #w will have the same number of cols as X has.
      self.b= 0
    
    if display_loss:
      loss= {}
    
    for i in tqdm_notebook(range(epochs),total=epochs,unit='epoch'):
      dw= 0
      db= 0
      
      for x,y in zip(X,Y):
        dw+= self.grad_w(x,y)
        db+= self.grad_b(x,y)
      self.w -= learning_rate*dw
      self.b -= learning_rate*db
      
     
      if display_loss:
        Y_pred= self.sigmoid(self.perceptron(X))
        loss[i]= mean_squared_error(Y_pred,Y)
    
    if display_loss:
      plt.plot(loss.values())
      plt.xlabel('Epochs')
      plt.ylabel('Mean Squared Error')
    
  def predict(self,X):
    Y_pred= []
    for x in X:
      y_pred= self.sigmoid(self.perceptron(x))
      Y_pred.append(y_pred)
    return np.array(Y_pred)
  
```

```
from sklearn.metrics import accuracy_score

sig= SigmoidNeuron()

sig.fit(X_scaled_train,y_scaled_train,epochs=5000,learning_rate=0.015,display_loss=True) #we will now be able to see the progress and time.
```
