# Functions


```
def square_the_num(num):
  """squares a number"""
  return num**2

square_the_num(int(input('Enter a number: ')))
```

```
def sum_of_squares_upto(num):
  sum_=0
  for i in range(num+1):
    sum_+=i**2
  return sum_

val= sum_of_squares_upto(int(input('Enter a number: ')))

print(val)
```

```
def squares_within_limit(stop_val,start_val=0):
  """Prints squares of value starting from start_val, and ending at stop_val"""
  for i in range(start_val,stop_val):
    print(i,i**2)
    
squares_within_limit(5,2)
```

```
help(squares_within_limit)
```

**Advantage of set can be seen when finding an element in a set:**

```
my_list=list(range(1000000))
tic= time.time()
print(98731 in my_list)
print(131591 in my_list)
print(1111111111 in my_list)
toc= time.time()
print('Elapsed time:', toc-tic)
```

```
my_set=set(range(1000000))
tic= time.time()
print(98731 in my_set)
print(131591 in my_set)
print(1111111111 in my_set)

toc= time.time()
print(f'time elapsed: {toc-tic}')
```

**We can see above that finding an element in "set" is faster than in list; This is because it does not iterate on the elements, but finds the element by computing a hash from the key**

_but iterating on a list is generally faster than on set_

# Python Classes

```
class MobilePhone:
  """Sample dummy class to illustrate how 'classes' work in python"""
  def __init__(self,brand_name,price,is_android=False):
    self.brand_name=brand_name
    self.price=price
    self.is_android=is_android
    self.rating=-1
    
  def __str__(self):
    return f'{self.brand_name} {self.is_android} {self.price}'
    
  def has_rating(self):
    return self.rating > -1, self.rating

new_phone= MobilePhone('Motorola',6000,True)
```

```
print(new_phone.brand_name)
new_phone.is_android
```

**We can change the value as: **

```
new_phone.price=new_phone.price+2000

print(new_phone.price)
```

```
new_phone.has_rating()
```

```
new_phone.rating=2.5

print(new_phone.has_rating())
```

```
MobilePhone.__doc__
```

**_Inheritance in python_**

```
class iPhone(MobilePhone):
  def __init__(self,name):
    MobilePhone.__init__(self,name,50000,"i don't know")  
    #it's better to keep the argument names of inherited class and parent class matching
```

```
new_iPhone= iPhone('iPhone 5')
print(new_iPhone.is_android)
print(new_iPhone.price)
print(new_iPhone.brand_name)   
```
